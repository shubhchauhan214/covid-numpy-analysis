```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
cities = ["Delhi", "Mumbai", "Chennai", "Kolkata", "Banglore", "Hyderabad", "Ahmedabad","Pune", "Jaiput", "Lucknow"]
num_people = 1000
total_people = num_people * len(cities)

np.random.seed(42)

ages = np.random.randint(10, 80, size=total_people)

infection_status = np.random.choice([0,1], size=total_people, p=[0.7, 0.3])

recovery_days = np.where(infection_status == 1, np.random.randint(5,21, size=total_people), 0)

temperature = np.where(infection_status == 1, np.random.normal(loc =101, scale = 1.5, size=total_people), 98.6)

city_data = np.repeat(cities, num_people)
```

```python
print("Total people:", total_people)
print("Total infected:", np.sum(infection_status))
print("Infection rate:", np.mean(infection_status) * 100, "%")
print("Average age of infected:", np.mean(ages[infection_status == 1]))
print("Average recovery days:", np.mean(recovery_days[infection_status == 1]))
```

```python
city_infection_counts = []

for city in cities:
    mask = city_data == city
    infected_count = np.sum(infection_status[mask])
    city_infection_counts.append(infected_count)
    
plt.figure(figsize=(10,6))
plt.bar(cities, city_infection_counts, color='tomato', zorder=3)
plt.xlabel("City")
plt.ylabel("Infected People")
plt.title("Covid-19 Infections per City")
plt.xticks(rotation=45)
plt.grid(True, zorder=0)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(ages[infection_status == 1], bins=20, color='skyblue', edgecolor='black')
plt.title("Age Distribution of Infected People")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(temperature[infection_status == 1], bins=25, color='orange', edgecolor='black', zorder=3)
plt.title("Body Temperature of Infected People")
plt.xlabel("Temperature (F)")
plt.ylabel("People Count")
plt.grid(True, zorder=0)
plt.show()
```

```python
city_rates = []

for city in cities:
    mask = city_data == city
    total = len(infection_status[mask])
    infected = np.sum(infection_status[mask])
    rate = infected / total
    city_rates.append(rate)
    
highest_rate = np.max(city_rates)
highest_city = cities[np.argmax(city_rates)]

print(f"Highest infection rate: {highest_rate*100:.2f}% in {highest_city}")
```
